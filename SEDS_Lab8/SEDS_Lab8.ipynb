#  Bayes' law
Use Bayes’ law to calculate the probability of getting a data science job if you’ve gotten an interview for the job. This could be written P(get the DS job | interview). You’ll have to use Bayesian probability methods (your intuition or beliefs) to assign values to the different components of Bayes’ law.

`Hint:`
* Usually the probability of getting a job from applying is very low, around 1%. This is based on experience and reading about how most people apply to 100 jobs in order to get 1 job. 
* P(interview | get job) is 99%. Almost no one will hire someone they didn't interview. 
* P(interview | !get job) will be set to 10%. We might interview 10% of candidates, and only one gets the job.

***Bayes' Theorem states:***
$$ 
P(A|B)=\frac{P(B|A)  P(A)}{P(B)} $$
**SO:**
$$
P(\text{get the DS job } | \text{ interview}) = \frac{P(\text{interview }|\text{ get the DS job })P(\text{ get the DS job })}{P(\text{ interview })}
$$
we have:

$ P(\text{get the DS job }) = 0.01 \\ $
$ P(\text{interview }|\text{ get the DS job }) = 0.99 \\ $
$ P(\text{interview }| \text{ not get the DS job }) = 0.1 \\ $
$ P(\text{interview }) =  P(\text{get the DS job })P(\text{interview }|\text{ get the DS job }) + P(\text{get the DS job })P(\text{interview }| \text{ not get the DS job }) $



```python
#Your Solution
p_getDsJob = 0.01
p_inter__getDsJob = 0.99
p_inter = 0.1*0.99 + 0.1*0.1

p_getDsJob__inter = (p_inter__getDsJob*p_getDsJob)/(p_inter)
p_getDsJob__inter
```

# MISO distributions
Determine the type of distribution and extract the distribution's PDF parameters from the MISO wind power generation data (the MWh column), stored under  data/miso_wind_data.csv. Write a short analysis explaining why you chose that distribution and what the parameters mean.

`Hint:`

```python
miso_data  = pd.read_csv('data/miso_wind_data.csv')
miso_data.sample(100)
```

```python
mwh_data = miso_data['MWh'].dropna().astype(float)
```

```python
miso_data.describe()
```

```python
sns.histplot(mwh_data, kde=True, color='b')
```

```python
import scipy.stats as stats
results = {}
```

### Test Normal Distribution
**Formula:**
$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^{2}}}
$$
**Parameters:**
- $\mu$ (mean): Determines the center of the distribution.
- $\sigma$ (standard deviation): Measures the spread or dispersion of the data.

```python
x = np.linspace(mwh_data.min(), mwh_data.max(), 100)
params = stats.norm.fit(mwh_data)
pdf = stats.norm.pdf(x, *params)
results["norm"] = (params, pdf)
plt.plot(x, pdf, label=f'Normal PDF', color='red')
plt.hist(mwh_data, bins=30, density=True, alpha=0.6, color='blue', label='Data')
plt.legend()
plt.title('Fitted Distributions')
plt.xlabel('MWh')
plt.ylabel('Density')
plt.show()
```

### Exponential Distribution
**Formula:**
$$ f(x;\lambda) = \lambda e^{-\lambda x}, \ \ x \ge  0 $$
**Parameters:**
- $\lambda$ (rate parameter): Inverse of the mean $(\lambda = 1 / \text{mean})$

```python
params = stats.expon.fit(mwh_data)
pdf = stats.expon.pdf(x, *params)
results["expon"] = (params, pdf)
plt.plot(x, pdf, label=f'Exponential PDF', color='red')
plt.hist(mwh_data, bins=30, density=True, alpha=0.6, color='blue', label='Data')
plt.legend()
plt.title('Fitted Distributions')
plt.xlabel('MWh')
plt.ylabel('Density')
plt.show()
```

### Test Weibull Distribution
**Formula:**
$$ f(x;k,\lambda) = \frac{k}{\lambda}\Big(\frac{x}{\lambda}\Big)^{k-1}e^{-\big(\frac{x}{\lambda}\big)^k}, \ \ x\ge 0 $$
**Test:**
- $k$ (shape parameter): Determines the shape of the distribution
  - $k<1$ Decreasing failure rate (e.g., early failures).
  - $k=1$: Constant failure rate (exponential distribution).
  - $k>1$: Increasing failure rate (e.g., aging or wear-out failures).
- $\lambda$ (scale parameter): Stretches or shrinks the distribution.

```python
params = stats.weibull_min.fit(mwh_data)
pdf = stats.weibull_min.pdf(x, *params)
results["weibull_min"] = (params, pdf)
plt.plot(x, pdf, label=f'Weibull PDF', color='red')
plt.hist(mwh_data, bins=30, density=True, alpha=0.6, color='blue', label='Data')
plt.legend()
plt.title('Fitted Distributions')
plt.xlabel('MWh')
plt.ylabel('Density')
plt.show()
```

### Test Gamma Distribution
**Formula:**
$$ f(x;k,\theta) = \frac{x^{k-1}e^{-x/\theta}}{\theta^k\Gamma(k)}, \ \ x\ge 0 $$
**Parameters:**
- $k$ (shape parameter): Determines the shape of the distribution.
- $θ$ (scale parameter): Scales the distribution horizontally.

```python
params = stats.gamma.fit(mwh_data)
pdf = stats.gamma.pdf(x, *params)
results["gamma"] = (params, pdf)
plt.plot(x, pdf, label=f'Gamma PDF', color='red')
plt.hist(mwh_data, bins=30, density=True, alpha=0.6, color='blue', label='Data')
plt.legend()
plt.title('Fitted Distributions')
plt.xlabel('MWh')
plt.ylabel('Density')
plt.show()
```

| Distribution      | Shape                   | Domain            | Common Use Case                           |
|-------------------|-------------------------|-------------------|-------------------------------------------|
| Normal            | Bell-shaped, symmetric | (-∞, ∞)           | Modeling symmetric data like heights.     |
| Exponential       | Decaying exponentially | [0, ∞)            | Waiting times (e.g., inter-arrival times).|
| Weibull           | Flexible (depends on k)| [0, ∞)            | Reliability or failure modeling.          |
| Gamma             | Skewed right           | [0, ∞)            | Waiting times with bursts or clusters.    |


```python
for name, (params, _) in results.items():
    ks_stat, p_value = stats.kstest(mwh_data, name.lower(), args=params)
    print(f'{name}: KS Statistic = {ks_stat}, P-value = {p_value}')
```


# Bootstrap Sampling
Apply bootstrap sampling to the MISO wind power generation data and plot the distribution of the mean wind power generation values.

```python
n_bootstrap = 1000  
bootstrap_means = []
for _ in range(n_bootstrap):
    sample = np.random.choice(mwh_data, size=len(mwh_data), replace=True)
    bootstrap_means.append(np.mean(sample))
    
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_means, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Bootstrap Means for Wind Power Generation')
plt.xlabel('Mean MWh')
plt.ylabel('Density')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
lower_ci = np.percentile(bootstrap_means, 2.5)
upper_ci = np.percentile(bootstrap_means, 97.5)

print(f"95% Confidence Interval for the Mean: ({lower_ci:.2f}, {upper_ci:.2f})")
```
